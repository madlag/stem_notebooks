```python
from skimage import data
from skimage.feature import Cascade

import matplotlib.pyplot as plt
from matplotlib import patches

# Load the trained file from the module root.
trained_file = data.lbp_frontal_face_cascade_filename()

# Initialize the detector cascade.
detector = Cascade(trained_file)

#img = data.astronaut()
from skimage import io
img = io.imread("riverdale.jpg")

detected = detector.detect_multi_scale(img=img,
                                       scale_factor=1.2,
                                       step_ratio=1,
                                       min_size=(60, 60),
                                       max_size=(250, 250))

plt.imshow(img)
img_desc = plt.gca()
plt.set_cmap('gray')
#print(detected)
colors = 'red,blue,green,white'.split(",")
print(colors)
index=0
for patch in detected:
    print(patch)
    img_desc.add_patch(
        patches.Rectangle(
            (patch['c'], patch['r']),
            patch['width'],
            patch['height'],
            fill=False,
            color=colors[index],
            linewidth=2
        )
    )
    index = index + 1
    

plt.show()
```

# Profondeur, Pression et Volume. Applications à la plongée.

```python
%matplotlib notebook

import matplotlib.pyplot as plt
```

```python
z_list = list([i for i in range(0, 101, 10 )])
z_list
```

```python
def pression_pour_z(z):
    return z/10+1
```

```python
p_list = list([pression_pour_z(z) for z in z_list])
p_list
```

```python
plt.plot(z_list, p_list, marker='+', linestyle='dashed')
plt.xlabel('profondeur (mètres)')
plt.ylabel('pression (bars)')
```

```python
pression_pour_z(z=2)
```

```python
def volume_pour_profondeur(V1, z):
    return V1/(z/10+1)
    
```

```python
volume_pour_profondeur(V1=2, z=3)
```

```python
v_list = list([volume_pour_profondeur(V1= 1, z=z) for z in z_list])
v_list
```

```python
plt.plot(z_list, v_list, marker='+', linestyle='dashed')
plt.xlabel('profondeur (mètres)')
plt.ylabel('volume (litres)')
```
